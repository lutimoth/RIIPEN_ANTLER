```python
def scatterplots(xcolumn,columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = xcolumn, y = col, data=pooldata, alpha=0.5)
        ax.set(xlabel=xcolumn, ylabel=col)
        ax.tick_params(axis='x', labelrotation=45)
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
scatterplots('time_stamp',colume, ncol=3, figsize=(15, 15))
```

```python
pooldatasummary=pooldata[columesummary].dropna()
pooldatasummary=pooldatasummary.sort_values(by='time_stamp')
```

```python
pooldatasummary.set_index('time_stamp',inplace=True)
```

```python
pooldatasummary.head()
```

```python
pooldatasummary.index[0]
```

```python
def summary(time,data,parameter,diff):
    df = data
    time = pd.to_datetime(time)
    time_diff= time-pd.Timedelta(diff)

    df_summary = df.loc[time_diff: time]


    df_summary.plot(y=parameter, use_index=True)
    max_parameter = max(df_summary[parameter])
    min_parameter = min(df_summary[parameter])
    avg_parameter = df_summary[parameter].mean()
    
    print(f """
            In the last {diff} from {time_diff.floor('S')} to {time.floor('S')} we have the following information:\n
            Total {parameter} measurements: {len(df_summary)}\n
            Max {parameter}: {max_parameter:.3f} \n 
            Min {parameter}: {min_parameter:.3f} \n
            Average {parameter}: {avg_parameter:.3f} \n
            """)
```

```python
pooldatasummary.plot(y='PH', use_index=True)
```

```python
summary('2022-11-07 21:59:24.885-04:00', pooldatasummary,'PH','2 days')
```

```python
summary('2022-11-07 21:59:24.885-04:00', pooldatasummary,'salinity','2 days')
```
